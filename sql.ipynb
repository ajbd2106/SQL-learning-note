6.colleges paired with highest gpa of their applicants???

```python
%%sql
Select distinct College.cName, state, GPA
From College, Apply, Student
Where College.cName = Apply.cName
And Apply.sID = Student.sID
And GPA >= all (select GPA from Student, Apply
Where Student.sID = Apply.sID
And Apply.cName = College.cName)
```

17.amount by which avg gpa of students applying to CS
exceeds avg gpa of students not applying to CS

```python
%%sql
select cs.avgGPA-noncs.avgGPA from 
(select avg(gpa) as avgGPA from student where sid in
 (select sid from apply where major='CS' ))as cs,
(select avg(gpa) as avgGPA from student where sid not in
 (select sid from apply where major='CS' ))as noncs;
```

# join :the default join in sql is inner join

1.students and their gpa who apply for MIT's CS

```python
%%sql
select sname,gpa from  student join apply  
on student.sid=apply.sid where cname='MIT' and
major='CS'
```

another way

```python
%%sql
select sname,gpa from  student join apply  using(sid) 
where cname='MIT' and
major='CS'
```

2.join three tables based on sid and cname

```python
%%sql
select * from (select * from apply join student on apply.sid=student.sid)as A
join college on A.cname=college.cname ;
```

3.left outer join: 
all students paired with the majors and schools they applied for.

```python
%%sql
select sname,sid,cname,major from 
student left outer join apply using(sid);
```

2.delete all students who applied to more than two different majors

```python
%%sql
delete from student 
where sid in(select sid from apply
             group by sid having count(distinct major)>2);
```

4.turn the highest ee applicants into cs applicants

```python
%%sql
update apply set major='CS'
where major='EE'
and sid in(select sid from student where gpa>=
           all(select gpa from student where sid in 
              (select sid from apply where major='EE')))
```
